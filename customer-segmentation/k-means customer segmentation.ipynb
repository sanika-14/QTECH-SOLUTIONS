```python
import pandas as pd

# Load dataset
data = pd.read_csv('marketing_campaign.csv') 
print(data.head())
```

```python
# Reload the dataset with the correct delimiter (tab-separated values)
data = pd.read_csv('marketing_campaign.csv', delimiter='\t')

# Display the first few rows and dataset info to understand its structure
data_info = data.info()
data_preview = data.head()

data_info, data_preview
```

```python
# Check for missing values in the dataset
missing_values = data.isnull().sum()

# Display columns with missing values
missing_values[missing_values > 0]
```

```python
# Impute missing values in the 'Income' column with the median value
median_income = data['Income'].median()
data['Income'].fillna(median_income, inplace=True)

# Verify there are no more missing values in the 'Income' column
data['Income'].isnull().sum()
```

```python
# Identify categorical columns for encoding
categorical_columns = ['Education', 'Marital_Status']

# Apply one-hot encoding to the categorical columns
data_encoded = pd.get_dummies(data, columns=categorical_columns, drop_first=True)

# Display the updated columns to confirm encoding
data_encoded.columns.tolist()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Select features for clustering, including one-hot encoded columns
feature_columns = [
    # Numerical features
    'Income', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts',
    'MntFishProducts', 'MntSweetProducts', 'MntGoldProds',
    'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases',
    'NumStorePurchases', 'NumWebVisitsMonth',
    # Education features (one-hot encoded)
    'Education_Basic', 'Education_Graduation', 'Education_Master', 'Education_PhD',
    # Marital Status features (one-hot encoded)
    'Marital_Status_Alone', 'Marital_Status_Divorced', 'Marital_Status_Married',
    'Marital_Status_Single', 'Marital_Status_Together', 'Marital_Status_Widow',
    'Marital_Status_YOLO'
]

# Prepare the features for clustering
X = data_encoded[feature_columns]

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine optimal number of clusters using elbow method and silhouette score
max_clusters = 10
inertias = []
silhouette_scores = []

for k in range(2, max_clusters + 1):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

# Plot elbow curve
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(range(2, max_clusters + 1), inertias, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')

# Plot silhouette scores
plt.subplot(1, 2, 2)
plt.plot(range(2, max_clusters + 1), silhouette_scores, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score Method')
plt.tight_layout()
plt.show()

# Select optimal number of clusters (let's say it's 4 based on the plots)
optimal_k = 4
final_kmeans = KMeans(n_clusters=optimal_k, random_state=42)
cluster_labels = final_kmeans.fit_predict(X_scaled)

# Add cluster labels to the original dataframe
data_encoded['Cluster'] = cluster_labels

# Analyze cluster characteristics
cluster_summary = data_encoded.groupby('Cluster')[feature_columns].mean()
print("\nCluster Centers (Mean Values):")
print(cluster_summary)

# Visualize clusters using PCA for dimensionality reduction
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)

# Create 3D scatter plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], 
                    c=cluster_labels, cmap='viridis')
plt.colorbar(scatter)
ax.set_xlabel('First Principal Component')
ax.set_ylabel('Second Principal Component')
ax.set_zlabel('Third Principal Component')
plt.title('Clusters Visualization using PCA')
plt.show()

# Calculate and display the size of each cluster
cluster_sizes = data_encoded['Cluster'].value_counts().sort_index()
print("\nCluster Sizes:")
print(cluster_sizes)

# Calculate and display educational distribution within each cluster
print("\nEducational Distribution by Cluster:")
education_cols = ['Education_Basic', 'Education_Graduation', 'Education_Master', 'Education_PhD']
education_dist = data_encoded.groupby('Cluster')[education_cols].mean()
print(education_dist)

# Calculate and display marital status distribution within each cluster
print("\nMarital Status Distribution by Cluster:")
marital_cols = ['Marital_Status_Alone', 'Marital_Status_Divorced', 'Marital_Status_Married',
                'Marital_Status_Single', 'Marital_Status_Together', 'Marital_Status_Widow',
                'Marital_Status_YOLO']
marital_dist = data_encoded.groupby('Cluster')[marital_cols].mean()
print(marital_dist)

# Calculate feature importance for each cluster
feature_importance = pd.DataFrame(
    final_kmeans.cluster_centers_,
    columns=feature_columns
)

print("\nFeature Importance by Cluster:")
print(feature_importance)

# Add a summary of key characteristics for each cluster
print("\nCluster Profiles:")
for cluster in range(optimal_k):
    print(f"\nCluster {cluster}:")
    # Get the most common education level
    edu_profile = education_dist.iloc[cluster]
    most_common_edu = edu_profile[edu_profile == edu_profile.max()].index[0].replace('Education_', '')
    
    # Get the most common marital status
    marital_profile = marital_dist.iloc[cluster]
    most_common_marital = marital_profile[marital_profile == marital_profile.max()].index[0].replace('Marital_Status_', '')
    
    # Get average income and spending
    avg_income = cluster_summary.loc[cluster, 'Income']
    total_spending = sum(cluster_summary.loc[cluster, [col for col in feature_columns if col.startswith('Mnt')]])
    
    print(f"- Predominant Education: {most_common_edu}")
    print(f"- Predominant Marital Status: {most_common_marital}")
    print(f"- Average Income: ${avg_income:,.2f}")
    print(f"- Average Total Spending: ${total_spending:,.2f}")
```

```python
# Calculate average response rate per cluster
cluster_response = data_encoded.groupby('Cluster')['Response'].mean()
print("\nResponse Rate by Cluster:")
print(cluster_response)

# Calculate average campaign acceptance by cluster
campaign_cols = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']
campaign_acceptance = data_encoded.groupby('Cluster')[campaign_cols].mean()
print("\nCampaign Acceptance Rates by Cluster:")
print(campaign_acceptance)
```

```python
# Cross-tabulate clusters with key metrics
def analyze_cluster_quality(data, cluster_col='Cluster'):
    # Calculate coefficient of variation for key metrics
    metrics = ['Income', 'MntWines', 'MntMeatProducts', 'NumWebPurchases']
    
    for metric in metrics:
        cv = data.groupby(cluster_col)[metric].agg(['mean', 'std']).assign(
            cv=lambda x: x['std'] / x['mean']
        )
        print(f"\nCoefficient of Variation for {metric}:")
        print(cv['cv'])

analyze_cluster_quality(data_encoded)
```

```python
# Save clustered data for future use
data_encoded.to_csv('customer_segments.csv')

# Save cluster labels and key statistics
cluster_stats = data_encoded.groupby('Cluster').agg({
    'Income': 'mean',
    'MntWines': 'mean',
    'MntMeatProducts': 'mean',
    'NumWebPurchases': 'mean',
    'Response': 'mean'
}).round(2)

cluster_stats.to_csv('cluster_statistics.csv')
```
